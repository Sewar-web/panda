```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('vgsales.csv')
```

Which company is the most common video game publisher?


```python
publisher = df.groupby('Publisher').count()
publisher.sort_values('Name',ascending=False)[['Name']].head(1)
```

What’s the most common platform?


```python
platform = df.groupby('Platform').count()
platform.sort_values('Name',ascending=False)[['Name']].head(1) 
```

What about the most common genre?

```python
genre = df.groupby('Genre').count()
genre.sort_values('Name',ascending=False)[['Name']].head(1)
```

What are the top 20 highest grossing games?

```python
global_sales=df.sort_values('Global_Sales',ascending=False)[['Name','Global_Sales']].head(20)
global_sales
```

For North American video game sales, what’s the median?

```python
# north=df.iloc['North American']
north = df['NA_Sales'].median()
north

```

For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
std = df['NA_Sales'].std()
mean = df['NA_Sales'].mean()
sales = df['NA_Sales'].iloc[0]
(sales-mean)/std
```

The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
wii_sales = df[df['Platform']=='Wii'].Global_Sales
print(wii_sales.mean())

not_wii_sales = df[df['Platform']!='Wii' ].Global_Sales
print(not_wii_sales.mean())
```
